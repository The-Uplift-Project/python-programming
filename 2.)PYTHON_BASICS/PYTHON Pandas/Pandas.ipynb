```python
#Impute nan val
df["sepal_length"].mean()
```

```python
df["sepal_length"].fillna(value="ABC") #it will put abc val in place of Nan
#their is also a function dropna() but we dont use here
```

```python
df['sepal_length'].fillna( value=df['sepal_length'].mean()  , inplace=True)
#replace nan value with its mean
```
